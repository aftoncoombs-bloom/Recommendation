```python
import graphlab
from datetime import timedelta
graphlab.canvas.set_target('ipynb')
```

```python
data = graphlab.SFrame.read_csv('transactions-comp.csv')
```

```python
analytics = graphlab.SFrame.read_csv('analytic_base_2015-2016.csv')
```

```python
analytics['tm_stamp'] = analytics['tm_stamp'].str_to_datetime('%Y-%m-%d %H:%M:%S')
```

```python
data['timestamp'] = data['timestamp'].str_to_datetime('%Y-%m-%d %H:%M:%S')
```

```python
i = analytics[17050]

print data[0]['timestamp']
print i['tm_stamp']
print i['one_time_trans_count']

delta = i['tm_stamp'] + timedelta(hours=6)
data_date_slice = data[(data['timestamp'] <= i['tm_stamp']) & (data['timestamp'] >= delta)]
print len(data_date_slice)
data_slice = data_date_slice[(data_date_slice['org'] == i['org']) & (data_date_slice['form'] == i['entity'])]
one_time_trans = data_slice[data_slice['recurring'] == 0]
one_time_trans_count = len(one_time_trans)
one_time_trans_vol = one_time_trans['donations_amt'].sum()
# update analytics values
i['one_time_trans_count'] = one_time_trans_count
i['one_time_trans_vol'] = one_time_trans_vol

print i['one_time_trans_count']
```
