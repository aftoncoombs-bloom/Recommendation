```python
df.sort_values('type')[['type', 'weight']].groupby('type').aggregate({'weight': [np.mean, np.std, np.var, stats.mode]})
```

```python
boxplot_data = [df[df.type==t].weight for t in df.sort_values('type').type.unique()]
```

```python
# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))
# Create an axes instance
ax = fig.add_subplot(111)
# Create the boxplot
bp = ax.boxplot(boxplot_data)
ax.set_xticklabels(df.sort_values('type').type.unique())
ax.set_ylim(80, 215)
```
