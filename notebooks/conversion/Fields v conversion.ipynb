```python
import graphlab
import graphlab.aggregate as agg
from datetime import timedelta
import datetime as dt
graphlab.canvas.set_target('browser')
```

## Data import & cleanup

```python
base = graphlab.SFrame.read_csv('analytic_base.csv')
qgiv_stats = graphlab.SFrame.read_csv('analytic_qgiv_stats.csv')

joined = base.join(qgiv_stats, on={'id':'base'})
```

```python
fields_conversion = joined['don_form_trans_count', 'don_form_trans_vol', 'opt_fields', 'req_fields', 'visits']
fields_conversion = fields_conversion[fields_conversion['visits'] != 0]
fields_conversion['conversion'] = fields_conversion['don_form_trans_count']/fields_conversion['visits'] * 100
```
