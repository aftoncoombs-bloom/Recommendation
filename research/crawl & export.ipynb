The notebooks can be converted to markdown documents with the command `jupyter nbconvert --no-input --to markdown`. We can build an easily referenced collection of information from the documents already created in the repository. We need to crawl the `/research/` directory, find all markdown files to copy and the notebook files to convert to markdown.

```python
import os
```

```python
copy_files = []
convert_files = []
for root, dirs, files in os.walk('./'):
    if root != './' and '.ipynb' not in root:
        #print(root)
        for f in files:
            #print("\t{}".format(f))
            
            if '.md' in f:
                copy_files.append("{}/{}".format(root, f))
            elif '.ipynb' in f:
                convert_files.append("{}/{}".format(root, f))
```

```python
print("Found {:,} markdown documents to copy".format(len(copy_files)))
print("Found {:,} python notebooks to convert".format(len(convert_files)))
```

```python
print("Markdown files:")
print(copy_files)
print()
print("Notebook files:")
print(convert_files)
```
