```python
import sys
sys.path.insert(1, '../../scripts/')
from s3_support import *

import pandas as pd
```

```python
q = '''select
            year,
            count(distinct(email)) as donor
        from transactions
        where status='A' and recurring=0
        group by year'''
ot = redshift_query_read(q, schema='production')
```

```python
q = '''select
            year,
            count(distinct(email)) as donor
        from transactions
        where status='A' and recurring!=0
        group by year'''
rec = redshift_query_read(q, schema='production')
```

```python
ot.columns = ['year', 'onetime']
rec.columns = ['year', 'recurring']
```

```python
df = ot.merge(rec, on='year')
df['ratio'] = df['onetime'] / df['recurring']
```
