```python
q = '''select 
            date_trunc('day', date) as day,
            sum(views) as views
        from production.ga
        where
            path ilike '%for%' and
            path not ilike '%receipt%' and
            path not ilike '%leaderboard%' and
            path not ilike '%control%' and
            path not ilike '%account%' and
            path not ilike '%cms%'
        group by date_trunc('day', date)'''
df_all_qgiv = redshift_query_read(q, schema='production')

print("qgiv traffic:")
print("{:,} entries".format(len(df_all_qgiv)))
print("{} to {}".format(df_all_qgiv['day'].min(), df_all_qgiv['day'].max()))
print("2024 average per day: {:,.2f}".format(df_all_qgiv[df_all_qgiv['day']>='2024-01-01']['views'].mean()))
```

```python
q = '''select
            date_trunc('day', date) as day,
            sum(views) as views
        from production.ga
        where
            path ilike '%event%' and
            path ilike '%donate%'
        group by date_trunc('day', date)'''
df_all_p2p = redshift_query_read(q, schema='production')

print("{:,} entries".format(len(df_all_p2p)))
print("{} to {}".format(df_all_p2p['day'].min(), df_all_p2p['day'].max()))
print("2024 average per day: {:,.2f}".format(df_all_p2p[df_all_p2p['day']>='2024-01-01']['views'].mean()))
```
