```python
import pandas as pd
import numpy as np

import sys
sys.path.insert(1, '../../scripts/')
from s3_support import *
```

```python
q = 'select count(id) as pageviews, url from pages group by url'
df = redshift_query_read(q, schema='secure')
```

```python
df['url_clean'] = df['url'].str.replace('\d+', '')
df.tail()
```

```python
df_grpd = df.groupby('url')['pageviews'].sum().reset_index()
df_grpd = df_grpd[~df_grpd['url'].str.startswith('file:')]
df_grpd['perc'] = df_grpd['pageviews'] / df_grpd['pageviews'].sum()
```
