# twitter

```python
tw = get_dataframe_from_file("qgiv-stats-data", "social.twitter.csv")
tw.tail(3)
```

```python
print("{} forms, {} orgs".format(len(tw['form'].unique()), len(tw['org'].unique())))
print("-"*40)
print("{:.2f} average tweets per form".format(tw.groupby('form')['content'].count().mean()))
print("{:.2f} average tweets per org".format(tw.groupby('org')['content'].count().mean()))
print("{:.2f} average likes; {:.2f} retweets".format(tw['like_count'].mean(), tw['retweet_count'].mean()))
```

# facebook

```python
fb = get_dataframe_from_file("qgiv-stats-data", "social.facebook.csv")
fb.tail(3)
```

```python
len_forms = len(fb['form'].unique())
len_orgs = len(fb['org'].unique())
print("{} forms, {} orgs".format(len_forms, len_orgs))
print("-"*40)
print("avg:")
print("\treaction_count: {:.2f}".format(fb['reaction_count'].mean()))
print("\tcomment_count: {:.2f}".format(fb['comment_count'].mean()))
print("\tshare_count: {:.2f}".format(fb['share_count'].mean()))
print("-"*40)

len_reactions = len(fb[fb['reaction_count']>0])
perc_reactions = (float(len_reactions) / float(len_forms)) * 100.
len_comments = len(fb[fb['comment_count']>0])
perc_comments = (float(len_comments) / float(len_forms)) * 100.
len_shares = len(fb[fb['share_count']>0])
perc_shares = (float(len_shares) / float(len_forms)) * 100.

print("{} ({:.0f}%) forms w/ > 0 reactions; {:.2f} avg".format(len_reactions, perc_reactions, fb[fb['reaction_count']>0]['reaction_count'].mean()))
print("{} ({:.0f}%) forms w/ > 0 comments; {:.2f} avg".format(len_comments, perc_comments, fb[fb['comment_count']>0]['comment_count'].mean()))
print("{} ({:.0f}%) forms w/ > 0 shares; {:.2f} avg".format(len_shares, perc_shares, fb[fb['share_count']>0]['share_count'].mean()))
```
