```python
import sys
sys.path.insert(1, "/home/ubuntu/recommendation/scripts/")
from s3_support import *
```

```python
q_1 = """select
            f.type,
            count(t.id) as trans_count,
            sum(t.amount) as trans_vol,
            sum(t.qgiv_fee) as trans_fees,
            count(distinct(t.form)) as active_forms,
            count(distinct(t.org)) as active_orgs
        from production.transactions as t
            left join production.form as f on t.form = f.id
        where
            t.year=2024 and
            t.status='A'
        group by f.type"""
df_1 = redshift_query_read(q_1, schema='production')

q_2 = """select
            date_part('month', t.date) as month,
            f.type,
            count(t.id) as trans_count,
            sum(t.amount) as trans_vol,
            count(distinct(t.form)) as active_forms,
            count(distinct(t.org)) as active_orgs
        from production.transactions as t
            left join production.form as f on t.form = f.id
        where
            t.year=2024 and
            t.status='A'
        group by f.type, date_part('month', t.date)"""
df_2 = redshift_query_read(q_2, schema='production')
```
