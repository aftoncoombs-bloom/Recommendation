```python
import requests, json
import pandas as pd

URL = "https://secure.qgiv.com/admin/qgivadmin/statistics/export_feature_adoption.php"
KEY = "DSQR59VwyFhw21PKDF4K"
```

# export

```python
payload = {'key': KEY}
rsp = requests.post(URL, data=payload)
rsp_obj = json.loads(rsp.text)
```

```python
rsp_obj[0]['implementation_data'].keys()
```

### storing integrations data

```python
int_data = rsp_obj[0]['implementation_data']['service_integrations']
```

```python
df_integrations = pd.DataFrame(int_data)
df_integrations.columns = ['count', 'org']
df_integrations.head()
```

```python
df_integrations.to_csv("integrations.csv", index=False)
```

### storing dl data

```python
df_dl = pd.DataFrame(rsp_obj[0]['bidirectional_adoption_data']['dl_onetime'])
df_dl.columns = ['onetime', 'form']
```

```python
df_rec = pd.DataFrame(rsp_obj[0]['bidirectional_adoption_data']['dl_recurring'])
df_rec.columns = ['recurring', 'form']
```

```python
df_dl = df_dl.merge(df_rec, on='form')
df_dl.head()
```

```python
df_dl.to_csv("dl_transactions.csv", index=False)
```
