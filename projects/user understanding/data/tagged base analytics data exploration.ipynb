```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("org_week_tagged_analytics.csv")
```

```python
# cleanup
df.fillna(0, inplace=True)
df['org'] = df['org'].astype('int')
```

```python
df.describe()
```

```python
datapoints = [c.replace('wk12_', '') for c in df.columns if 'wk12' in c]
datapoints = list(set(datapoints))
week_steps = [1, 4, 12, 24, 52]
```

```python
key = "{}_{}"
df_data = []
for datapoint in datapoints:
    data = {'key': datapoint}
    for week_step in ['wk1', 'wk4', 'wk12', 'wk24', 'wk52']:
        data[week_step] = df[key.format(week_step, datapoint)].mean()
    df_data.append(data)
df_steps = pd.DataFrame(df_data)
```

```python
df_steps['wk4_perc'] = (df_steps['wk4'] - df_steps['wk1']) / df_steps['wk1']
df_steps['wk12_perc'] = (df_steps['wk12'] - df_steps['wk4']) / df_steps['wk4']
df_steps['wk24_perc'] = (df_steps['wk24'] - df_steps['wk12']) / df_steps['wk12']
df_steps['wk52_perc'] = (df_steps['wk52'] - df_steps['wk24']) / df_steps['wk24']

print("Looking at rate of growth")
df_steps[['key', 'wk4_perc', 'wk12_perc', 'wk24_perc', 'wk52_perc']]
```
