The primary idea is to create a simple function API through which a feature for a given form can be altered to determine the degree to which it influences conversion. Ideally, this is something that could be run across many forms either simultanesoulsy or in rapid iteration in order to identify maximum ROI features.

Presumed components:

1. Embedding: possible targets are org, segment, region
2. Trained models targeting conversion, average transaction amount, onetime/recurring ratio

Top level functions

- form_feature_optimization(form: int, feature: str)
- form_optimization(form: int)
