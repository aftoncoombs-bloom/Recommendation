```python
rolling_means = transactions \
    .groupby(['form', 'date']) \
    ['donations_amt'] \
    .sum() \
    .rolling(window=7) \
    .mean() \
    .reset_index() \
    .rename(columns={'donations_amt': 'previous_seven_day_mean'})
```
