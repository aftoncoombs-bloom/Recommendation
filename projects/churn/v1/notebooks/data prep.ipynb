```python
orgs_new_download = pd.read_csv("organizations_download.csv")
orgs = pd.read_csv("organizations.csv")
orgs_clean = pd.read_csv("organizations.clean.csv")
```

```python
print("clean orgs:")
print("{} rows, {} columns".format(len(orgs_clean), len(orgs_clean.columns)))
orgs_clean.columns
```

```python
# clean up column names
old_column_names = orgs_new_download.columns
for c in old_column_names:
    new_column = c.replace(" ", "_").lower()
    if new_column == 'go_live_date':
        new_column = 'live_date'

    orgs_new_download[new_column] = orgs_new_download[c]
    orgs_new_download.drop(c, axis=1, inplace=True)
```

```python
# drop unnecessary columns from new download
cleaned_cols = orgs_clean.columns.tolist()
cleaned_new_download = orgs_new_download[cleaned_cols]

print("{} rows, {} columns".format(len(cleaned_new_download), len(cleaned_new_download.columns)))
```

```python
# store cleaned up download
orgs_new_download.to_csv("organizations.clean.csv", index=False)
```
