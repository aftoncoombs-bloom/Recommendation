```python
import pandas as pd

import sys
sys.path.insert(1, '../../../scripts/')
from s3_support import *
```

# Load IRS Data

```python
irs_data = get_dataframe_from_file('tax-info', 'irs_990.csv')
irs_data['source'] = 'IRS'
```

```python
irs_data
```

# Load OpenData Data

```python
open_data = get_dataframe_from_file('tax-info', 'opendata_990.csv')
open_data['source'] = 'Open Data'
```

```python
open_data
```

# Load ProPublica Data

```python
pro_publica = get_dataframe_from_file('tax-info', 'propublica_990.csv')
pro_publica['source'] = 'Pro Publica'
```

```python
pro_publica
```

# Combine 990 Data

```python
combined_data = pd.concat([irs_data, open_data, pro_publica], sort=False)
```

```python
# Sanity check to make sure all the data was combined properly
assert(len(combined_data) == (len(irs_data) + len(open_data) + len(pro_publica)))
```

```python
combined_data
save_dataframe_to_file('tax-info', 'org_revenue_data.csv', combined_data)
```
