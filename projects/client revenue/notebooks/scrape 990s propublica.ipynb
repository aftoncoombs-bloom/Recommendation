Exploring the [ProPublica](https://projects.propublica.org/nonprofits/api) nonprofit explorer API.

```python
import requests, json
import pandas as pd
```

## Generic info

```python
url = "https://projects.propublica.org/nonprofits/api/v2/search.json?q={}"
```

```python
rsp = requests.get(url.format("propublica"))
json.loads(rsp.content)
```

## EIN

```python
url = "https://projects.propublica.org/nonprofits/api/v2/organizations/{}.json"
```

```python
rsp = requests.get(url.format("142007220"))
data = json.loads(rsp.content)
```

```python
for filing in data['filings_with_data']:
    print("{}: {}".format(filing['tax_prd_yr'], filing['totrevenue']))
```
