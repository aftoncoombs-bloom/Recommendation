```python
import pandas as pd
import numpy as np
```

```python
print("load & prep data")
print("\treading CSV's")
df_base = pd.read_csv("~/Repositories/datasets/analytics/a_base.csv")
df_qgiv = pd.read_csv("~/Repositories/datasets/analytics/analytics_qgiv.csv")
df_qgiv_base = pd.read_csv("~/Repositories/datasets/analytics/a_qgiv_base_id.csv")

print("\tmerge dataframes")
df_qgiv = df_qgiv.merge(df_qgiv_base, left_on="id", right_on="id_x")
df = df_qgiv.merge(df_base, left_on="base", right_on="id", how="left")

print("\tadding conversion & embedding columns")
df['conversion'] = df['don_form_trans_count'] / (df['visits']+df['mobile_visits'])
df['conversion'].fillna(0., inplace=True)
df['conversion'].replace(np.inf, 1., inplace=True)
```

```python
org_dl_means = df.groupby('org_x')['enable_donorlogins'].mean().reset_index()
org_dl_means.columns = ['org', 'dl']
org_dl_means.head()
```

```python
orgs_dl_changed = org_dl_means[(org_dl_means.dl!=1.)&(org_dl_means.dl!=0.)]
print("{} total orgs, {} orgs that changed dl setting".format(len(df.org_x.unique()), len(orgs_dl_changed)))
```

```python
# add column for whether or not the org has changed the setting
df['dl_setting_changed'] = df.org_x.isin(orgs_dl_changed.org.values)
```

```python
# add column for the last donor login setting
last_dl_settings = df.groupby('org_x')['enable_donorlogins'].last().reset_index()
df['dl_setting_last'] = df['org_x'].apply(lambda x: last_dl_settings[last_dl_settings['org_x']==x]['enable_donorlogins'].values[0])
df[['dl_setting_changed', 'dl_setting_last']].head()
```

```python
cols = ['new_rec_count', 'dl_new_rec_count', 'one_time_trans_vol', 'one_time_trans_count', 'kiosk_trans_vol', 'kiosk_trans_count', 'rec_trans_vol', 'rec_trans_count']
dl_on = df[df.enable_donorlogins==True]
dl_off = df[df.enable_donorlogins==False]

print("| Column | On | Off |")
print("|---|---|---|")
for c in cols:
    print("| {} | {:.6f} | {:.6f} |".format(c, dl_on[c].mean(), dl_off[c].mean()))
```

```python
trans = pd.read_csv("/Users/jeremyvanvalkenburg/Repositories/datasets/transactions/transactions_agg.csv")
```

```python
dl_on_trans = trans[trans.org.isin(orgs_dl_on)]
dl_off_trans = trans[~trans.org.isin(orgs_dl_on)]

print("{} transactions for on, {} transactions for off".format(len(dl_on_trans), len(dl_off_trans)))
print("${:.2f} mean transaction for on, ${:.2f} mean transaction for off".format(dl_on_trans.amount.mean(), dl_off_trans.amount.mean()))
print("${:.2f} average funds raised for dl on".format(dl_on_trans.amount.sum() / float(len(orgs_dl_on))))
print("${:.2f} average funds raised for dl off".format(dl_off_trans.amount.sum() / float(len(dl_off_trans.org.unique()))))
```
